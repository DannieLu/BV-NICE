## Compare results of BV-NICE and BART 

```python

n_rep = 20      # 0-19
n_folder = 77   # 1-77

pehe_klm = np.zeros([n_rep*n_folder,4])

i=0
for folder_ind in range(1 , n_folder+1 ):
    for file_ind in range(n_rep):
        
        dtemp = np.load("./results_bnice_v4_acic/" +str(folder_ind)+"_" + str(file_ind)+"_results.npy")
        rmse_min_ind_out = np.argmin( dtemp[:,2,3] )  # find the run where rmse is minimum
        pehe_klm[i,:] = [folder_ind, file_ind, dtemp[rmse_min_ind_out,2,0]/dtemp[rmse_min_ind_out,0,4], dtemp[rmse_min_ind_out,0,4] ]

        i+=1
    
```

```python
pehe_klm_ = pd.DataFrame({"Exp_id":pehe_klm[:,0],"Rep_id":pehe_klm[:,1], \
                          "PEHE":pehe_klm[:,2],"Model": np.repeat("BV-NICE",n_rep*n_folder) })
ax = sns.lineplot(x="Exp_id", y="PEHE", data=pehe_klm_)
print(pehe_klm_)
_=plt.plot()
```

```python
pehe_bart = np.zeros([n_rep*n_folder,6])

i=0
for folder_ind in range(1 , n_folder+1 ):
    for file_ind in range(n_rep):
        pehe_bart[i,:] = np.load("./results_bart_acic/" +str(folder_ind)+"_" + str(file_ind)+"_results.npy")
        i+=1
    
```

```python
pehe_bart
```

```python

pehe_bart_ = pd.DataFrame({"Exp_id":pehe_bart[:,0],"Rep_id":pehe_bart[:,1],\
                           "PEHE":pehe_bart[:,4]/pehe_bart[:,5],"Model": np.repeat("BART",n_rep*n_folder)})
ax = sns.lineplot(x="Exp_id", y="PEHE", data=pehe_bart_)
print(pehe_bart_)
_=plt.plot()
```

```python
pehe_klm_bart_ = pehe_klm_.append(pehe_bart_)
print( pehe_klm_bart_)
```

```python
_ = sns.lineplot(x="Exp_id", y="PEHE", hue="Model", data=pehe_klm_bart_)
_ = plt.title("ACIC2016")
plt.savefig('ACIC2016.png')
```
