```python


def load_jobs(filepath='./data/',istrain=True):
    
    if istrain:
        data_file = filepath+'jobs_DW_bin.train.npz'
    else:
        data_file = filepath+'jobs_DW_bin.test.npz'
        
    data = np.load(data_file)
    
    x = data['x']  # confounders
    y = data['yf'] # factual observation
    t = data['t']  # treatment
    e = data['e']  # randomized trial
    
    return x,y,t,e

def eval_Rpol(tau_hat,T,E,Y,lam=0):
    
    '''
    Evaluating policy risk.
    '''
    
    P = tau_hat>lam
    
    ind1 = np.logical_and(np.logical_and(P,T==1),E==1)
    ind0 = np.logical_and(np.logical_and(P==False,T==0),E==1)
    
    ind1a = np.logical_and(P==True,E==1)
    ind0a = np.logical_and(P==False,E==1)
    
    EY1 = np.mean(Y[ind1])
    EY0 = np.mean(Y[ind0])
    
    p1 = np.float(np.sum(ind1a))/np.sum(E)
    p0 = np.float(np.sum(ind0a))/np.sum(E)
    
    PR = 1 - (EY1*p1+EY0*p0)
    
    return PR


def compute_ATT(Y,T,E):
    
    t_ind = T==1
    ce_ind = np.logical_and(T==0,E==1)
    
    # n_t = np.sum(t_ind)
    # n_ce = np.sum(ce_ind)
    
    att = np.mean(Y[t_ind]) - np.mean(Y[ce_ind])
    
    return att

def compute_att_err(tau_full_hat,Y_full,T_full,E_full):
    
    att_ = compute_ATT(Y_full,T_full,E_full)
    
    att_hat = np.mean(tau_full_hat[T_full==1])
    
    return np.abs(att_-att_hat)


def onehot(t,dim):
    
    m_samples = t.shape[0]
    tt = np.zeros([m_samples,dim])
    
    for i in range(m_samples):
        tt[i,np.int(t[i])] = 1
        
    return tt
```

```python
def Pol_risk(tau_full_hat, model=""):
    lam_list = np.sort(tau_full_hat[E_full==1])
    lam_list = lam_list[::26]

    res = []
    pt = []

    for lam in lam_list:
        res.append(eval_Rpol(tau_full_hat,T_full,E_full,Y_full,lam))
        pt.append(np.sum(np.logical_and(tau_full_hat>lam,E_full==1))/np.float(np.sum(E_full)))

    res = np.array(res[-1::-1])
    pt = np.array(pt[-1::-1])
    
    out = pd.DataFrame({"Pol_risk": res, "Inc_rate": pt, "Model": np.repeat(model, len(res)) })
    
    return out
```

```python
pr_bnice = Pol_risk(tau_hat_bnice[0],model="BV-NICE")
pr_cfr = Pol_risk(tau_hat_cfr[0].reshape([-1,]),model="CFR")
pr_crf = Pol_risk(tau_hat_crf[0],model="Causal-RF")
pr_lreg = Pol_risk(tau_hat_lreg[0],model="Logistic-Reg")
```
